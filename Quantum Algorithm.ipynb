```python
from qiskit import *
%matplotlib inline
```

```python
## Deutsch Jozsa algorithm

## Define Blackbox
blackbox = '0000'
##

circuit = QuantumCircuit(len(blackbox)+1,len(blackbox))
circuit.x(len(blackbox))
circuit.h(range(len(blackbox)+1))
circuit.barrier()

## create blackbox
for index, value in enumerate(blackbox):
    if int(value):
        circuit.cx(index,len(blackbox))
##

circuit.barrier()
circuit.h(range(len(blackbox)+1))
circuit.x(len(blackbox))
circuit.measure(range(len(blackbox)), range(len(blackbox)-1,-1,-1))

simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend = simulator, shots = 1024).result()
counts = result.get_counts()
if list(counts.keys())[0].find('1') == -1:
    print('This function is constant.')
else:
    print('This function is balance.')
    
circuit.draw(output = 'mpl')
```
